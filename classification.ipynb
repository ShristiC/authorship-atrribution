```python
import nltk
# nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd
import numpy as np
```

```python
emma = gutenberg.raw('austen-emma.txt')
```

```python
text = {}
for i in gutenberg.fileids():
    author = i.split("-")[0]
    if author not in text:
        text[author] = []
    text[author].append(gutenberg.raw(i))
```

```python
authors = list(text.keys())
print(authors)
```

```python
df = pd.DataFrame(list(text.items()),columns = ['author','text'])
```

```python
df
```

```python
df2 = pd.DataFrame()

author = []
text = []
label = []

for i in df.iterrows():
    for j in range(len(i[1]['text'])):
        author.append(i[1]['author'])
        text.append(i[1]['text'][j])
        label.append(authors.index(i[1]['author']))
        
df2 = pd.DataFrame({'author': author, 'text': text, 'label': label})
df2
```
